### Introduction
* Regression discontinuity is a quasi-experimental technique to identify causal effects exploiting threshold-based assignment to a given treatment.
* Treatment is assigned based on a continuous variable called the assignment variable (or running variable, or forcing variable).
  * If an individual falls on one side of the cutoff, they (may) receive treatment
  * If an individual falls on the other side of the cutoff, (likely) no treatment
* As for IV and experiments in general, assignment to treatment and treatment status in RD can differ in practice!
  * With sharp RD, treatment status changes sharply at the threshold
    * Moving across the threshold changes the probability of treatment from 0% to 100%
  * With fuzzy RD, only the probability of treatment changes at the threshold
    * In this case, we can apply IV logic to account for non-compliance

Basic RD idea: Check for a noticeable difference (discontinuity) in the outcome between nearly identical people who fall on either side of arbitrary threshold.

* Two key practical conditions for an RD analysis:
  * Assignment variable must be continuous (unlike in typical RCTs)
  * Assignment variable must be measured before treatment (i.e., not caused or impacted by the treatment).
* If you want to argue that the RD estimate is the causal effect, you need two additional assumptions:
1. Covariate balance
  * No other treatment or covariate break happens at the same threshold.
2. No manipulation
  * The individual cannot be able to 'manipulate' the value of the assignment variable precisely (i.e., no self-selection: assignment is not manipulable).

### Regression discontinuity (RD): examples
* Assignment variable can be anything defined prior to treatment.
* Examples include:
  * Assigning medical treatment on the basis of illness severity
  * Newborns that weigh less than 1500 grams receive special care
  * Assigning social welfare benefits on the basis of income
  * If you income is below 10000USD per year, you get the program
  * Election victory based on the share of votes obtained
  * Receiving majority of votes (50\%) decides who assumes office

### Regression discontinuity (RD): intuition
* The precise cutoff point for any given program is, ultimately, an arbitrary decision: around the cutoff, treated and control groups are comparable along relevant criteria
  * Baby that weighs 1501 grams is roughly as underweight as one that weighs 1499 grams (i.e., should have similar rate of survival, all things equal)
* The only thing that differs is that those on one side get the treatment and those on the other don't
  * Baby that weighs 1499 grams goes to NICU, the baby that weights 1501 grams does not
* Thus, any difference we observe in outcomes just above and below the threshold, we can attribute to the impact of the policy
* Key assumption for causality: individuals cannot perfectly control which side of the cutoff they are on

  * Other covariates/baseline characteristics should not have discontinuities at the cutoff
2. Density test: individuals should not be able to sort themselves around the cutoff
  * If they do, then individuals are self-selecting into each group and continuity is violated

### External validity and LATE
* Instrumental variables and regression discontinuity have a common property:

**The estimates produced have a marginal interpretation, i.e., they are driven by a subset of the sample studied (LATE).**

* The estimate of $\beta_1$ has a marginal interpretation because it is generated by students near the eligibility threshold.
* Important to ask whether causal impacts would be different for individuals far from the threshold.
* Would merit aid affect lower scoring students differently?

### Sharp vs. Fuzzy RD
* Sharp RD: treatment status changes from 0\% to 100\% across the threshold.
* Fuzzy RD: treatment probability changes across the threshold, but by less than $100 \%$.
* In fuzzy RD, the threshold is an instrumental variable, a source of variation in treatment probability (i.e. similarly to the case of imperfect compliance of $\mathrm{RCT})$.
* So we can have RD versions of the first stage and reduced form!

### Takeaways
* Regression discontinuity is a powerful tool for causal inference.
  * It allows comparison of two groups who are nearly identical except for the fact they fall on different sides of an arbitrary threshold.
  * It's as close to an RCT as a non-RCT can get (if assumptions hold)
* Many public policies involve eligibility thresholds!
* If you are convinced that those on both sides of the policy threshold are similar but for the policy treatment itself, RD can provide causal estimates.
  * RD estimates are LATEs driven by those near the threshold.
* In sharp RD, the threshold determines 100\% of treatment status (i.e. below = no treatment, above $=$ treatment).
* In fuzzy RD, the threshold changes the probability of treatment, but not from 0 to $100 \%$
  * In this case, you can use IV logic to 're-scale' the reduced form effect of the RD with its first stage impact, just as we did in experiments with imperfect compliance.
